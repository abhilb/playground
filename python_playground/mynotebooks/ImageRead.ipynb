# Read an image and display it in Python.
> A quick how-to for my own reference (hopefully will be of interest to others too, hence thought of having it as blog post rather than a private gist) on reading an image and to display it in Python. Also get the data as a numpy array for further manipulation. I have also tried to time the different approaches to get an idea about the performance of each approach.

There are many ways to read an image and display it using python.

* Using **pillow** package
* Using **opencv** package
* Using **scipy** package

To display the image I have used the **matplotlib** package.

*Image source: pixabay.com*

-----


```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img_cv2 = cv2.imread('snake.png')

plt.imshow(img_cv2)
plt.show()
```

```python
from scipy import misc
img_scipy = misc.imread('snake.png')

plt.imshow(img_scipy)
plt.show()
```

```python
from PIL import Image
img_pil = Image.open('python-logo.png')

plt.imshow(img_pil)
plt.show()
```
