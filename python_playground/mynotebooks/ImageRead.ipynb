```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img_cv2 = cv2.imread('python-logo.png')

plt.imshow(img_cv2)
plt.show()
```

```python
from scipy import misc
img_scipy = misc.imread('python-logo.png')

plt.imshow(img_scipy)
plt.show()
```

```python
from PIL import Image
img_pil = Image.open('python-logo.png')

plt.imshow(img_pil)
plt.show()
```
