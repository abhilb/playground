```python
%matplotlib inline
import scipy.ndimage
import scipy.misc
import matplotlib.pyplot as plt


face = scipy.misc.face(gray=True)
face.shape
```

```python
plt.imshow(face, cmap='gray')
plt.show()
```

```python
gaussian = scipy.ndimage.filters.gaussian_filter(face, 3.0)
plt.imshow(gaussian, cmap='gray')
plt.show()
```
