```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from scipy import ndimage
from scipy import misc

face = misc.face(gray=True)
plt.imshow(face, cmap='gray')
plt.show()
```

```python
w,h = face.shape

w,h
```

```python
from numpy.lib import stride_tricks as st
strided_image = st.as_strided(face, shape=[w-14,h-14,15,15], strides=face.strides+face.strides)
%time means = strided_image.mean(axis=(2,3))
```

```python
plt.imshow(means, cmap='gray')
plt.show()
```
